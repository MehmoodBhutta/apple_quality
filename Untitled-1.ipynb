# ML 

```python
X = df.drop(['Quality', 'A_id', 'Acidity'], axis=1)
y = df['Quality']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

scale = ['Size', 'Weight', 'Sweetness', 'Crunchiness', 'Juiciness',
       'Ripeness']
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('scale', StandardScaler(), scale)
    ], remainder='passthrough')
```

```python
!pip install xgboost
```

```python
from xgboost import XGBClassifier

pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('xgb', XGBClassifier())
])
```

```python
label = LabelEncoder()

y_train_transfrom = label.fit_transform(y_train)
y_test_transfrom = label.transform(y_test)
```

```python
pipeline.fit(X_train, y_train_transfrom)
```

```python
accuracy = pipeline.score(X_test, y_test_transfrom)
print(f'Model Accuracy: {accuracy}')
```
